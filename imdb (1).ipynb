# importing the required libraries

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

url = https://www.imdb.com/search/title/?title_type=feature&primary_language=hi&sort=num_votes,desc&start=51&ref_=adv_nxtusing 
using this url we extract movie data

# declaring the list of empty variables, So that we can append the data overall

```python
Title = []
Date_Release = []
movieGenre = []
cast_star = []
time=[]
IMDBrating=[]
metascore =[]
IMDBvotes = []
gross = []
plot_summary = []
Director =[]
```

# creating an array of values and passing it in the url for dynamic webpages

```python
pages = np.arange(1,4000,50)
```

# the find the data from multiple webpages

```python
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?title_type=feature&primary_language=hi&sort=num_votes,desc&start="+str(page)+"&ref_=adv_nxt")
    soup = BeautifulSoup(page.text, 'html.parser')
    movie_data = soup.findAll('div', attrs = {'class': 'lister-item mode-advanced'})
    
    for store in movie_data:
        name = store.h3.a.text
        Title.append(name)                       #scraching the movie names 
        
        year_of_release = store.h3.find('span', class_ = "lister-item-year text-muted unbold").text
        Date_Release.append(year_of_release)      #scraching the movie release year  
        
       
        rate = store.find('div', class_ = "inline-block ratings-imdb-rating").text.replace('\n', '')
        IMDBrating.append(rate)                 #collecting imdb rating data
        
        meta = store.find('span', class_ = "metascore").text if store.find('span', class_ = "metascore") else "****"
        metascore.append(meta)                     #scraching metascore and if-else used for same movies don't mention metascore
        
        
        value = store.find_all('span', attrs = {'name': "nv"})
        
        vote = value[0].text
        IMDBvotes.append(vote)                   #scraching the movie votes
        
        grosses = value[1].text if len(value)>1 else '%^%^%^'
        gross.append(grosses)                     #collecting gross collection ata  
        
        describe = store.find_all('p', class_ = 'text-muted')
        description_ = describe[1].text.replace('\n', '') if len(describe) >1 else '*****'
        plot_summary.append(description_)           #scraching the movie plot summary  
        
        Gener = store.p.find("span", class_ = 'genre').text.replace('\n', '') if store.p.find('span', class_ = "genre") else "****"
        movieGenre.append(Gener)                     #scraching the movie genre to find which movie is dramaaction,comedy etc
        
        runtime = store.p.find("span", class_ = 'runtime').text.replace("min","") if store.p.find('span', class_ = "runtime") else "****"
        time.append(runtime)                         #scraching the movies duration in minutes
        
        cast = store.find("p", class_ = '')
        cast = cast.text.replace('\n', '').split('|')   # find the star and directors name
        cast = [x.strip() for x in cast]
        cast = [cast[i].replace(j, "") for i,j in enumerate(["Director:","Stars:"])] if len(cast)>1 else '%^%^%^'
        Director.append(cast[0])
        cast_star.append([x.strip() for x in cast[1].split(",")]) if len(cast) >1 else '*****'
        
```

# Make the Dataframe

```python
movie_DF = pd.DataFrame({'Title': Title, 'Year_relase': Date_Release, 'Watchtime': time, 'Movie_imdb_Rating': IMDBrating, 'Metascore': metascore,'Movie_votes':IMDBvotes,'Collection':gross,'plot_summary':plot_summary,'Director':Director,'cast':cast_star,"Gener":movieGenre})
```
